# Cluster overlap between training and test clusters 

### Import modules

```python
import math
from collections import Counter
```

### Define functions

```python
def common(set1, set2):
    intersect_Counter = Counter(set1) & Counter(set2)
    d = []
    for k, v in intersect_Counter.items():
        for i in range(v):
            d.append(k)

    return d
```

#### 1. Cosine score

```python
def cosine_score(set1, set2):

    return float(len(common(set1, set2))) / (math.sqrt(float(len(set1))) *
                                          math.sqrt(float(len(set2))))
```

#### 2. Normalized Jaccard score

```python
def jaccard_score(set1, set2):
    return float(len(common(set1, set2))) / float(len(set1 + set2)) 
```

#### 3. Overlap score

```python
def overlap_score(set1, set2):
    return float(len(common(set1, set2))) / min(len(set1), len(set2))
```

#### 4. Dice score

```python
def dice_score(set1, set2):
    return 2.0 * float(len(common(set1, set2))) / float(len(set1) + len(set2))
```

### Example of Cluster Overlap coefficient

```python
a = [1, 1, 2, 2]
b = [3, 3, 4, 5, 5, 6]

print(cosine_score(a, b))
print(jaccard_score(a, b))
print(overlap_score(a, b))
print(dice_score(a, b))
```

```python
a = [1, 1, 1, 1]
b = [1, 1, 1, 1]

print(cosine_score(a, b))
print(jaccard_score(a, b))
print(overlap_score(a, b))
print(dice_score(a, b))
```

```python
a = [1, 1, 2, 2]
b = [1, 1, 2, 2]

print(cosine_score(a, b))
print(jaccard_score(a, b))
print(overlap_score(a, b))
print(dice_score(a, b))
```

```python
a = [1, 1, 2, 2]
b = [1, 1]

print(cosine_score(a, b))
print(jaccard_score(a, b))
print(overlap_score(a, b))
print(dice_score(a, b))
```
