## eSEN

```python
from ase.io import read
from fairchem.core import FAIRChemCalculator
from fairchem.core.units.mlip_unit import load_predict_unit
```

### Download pretrained model's weight

- For eSEN:

    Go to https://huggingface.co/facebook/OMol25 and download, e.g., `esen_sm_direct_all.pt`

---

### Initialize MLIP models

```python
esen_predictor = load_predict_unit(
    path="esen_sm_direct_all.pt",
    device="cpu",
)

esen_calculator = FAIRChemCalculator(esen_predictor)
```

```python
atoms = read('benzene.xyz')  # or create ASE Atoms however you want
atoms.info = {"charge": 0, "spin": 1}
```

### Compute energy

```python
atoms.calc = esen_calculator
energy_eV = atoms.get_potential_energy()
print(energy_eV) # -6319.18481262752
```
