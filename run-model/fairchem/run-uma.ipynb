## UMA

UMA inference (ref: https://www.rowansci.com/blog/how-to-run-open-molecules-2025)

```python
from ase.io import read
from fairchem.core import FAIRChemCalculator
from fairchem.core.units.mlip_unit import load_predict_unit
```

### Download pretrained model's weight

- For UMA:

    Go to https://huggingface.co/facebook/UMA and downlaod `uma-s-1p2.pt`

---

### Initialize MLIP models

```python
uma_predictor = load_predict_unit(
    path="uma-s-1p2.pt",
    # device="cuda",
    inference_settings="turbo", # optional ("turbo" accelerates inference but locks predictor to first system size it sees)
)
uma_calculator = FAIRChemCalculator(
    uma_predictor,
    task_name="omol", # options: "omol", "omat", "odac", "oc20", "omc"
)
```

```python
atoms = read('benzene.xyz')  # or create ASE Atoms however you want
atoms.info = {"charge": 0, "spin": 1}
```

### Compute energy

#### UMA

```python
atoms.calc = uma_calculator
energy_eV = atoms.get_potential_energy()
print(energy_eV) # -6319.18481262752
```
